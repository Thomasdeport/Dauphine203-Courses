```python
import hashlib

def hash_password(text):
    hash = 0 
    for char in text : 
        hash = (hash * 281 ^ord(char) * 997) & 0xFFFFFFFF
    return hash
```

```python
from itertools import product
import string

def find_password(target_hash):


    chars = [chr(i) for i in range(32,128)]
    
    for length in range(4,5):  
        for attempt in product(chars, repeat=length):
            attempt_str = ''.join(attempt)
            if hash_password(attempt_str) == target_hash:
                return attempt_str
    return None

a = find_password(2868533088)

print(a)
```
